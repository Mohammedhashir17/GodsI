```python
import matplotlib.pyplot as plt

models = ['CNN (Proposed)', 'Haar Cascades', 'HOG']
accuracy = [94, 78, 85]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracy, color=['green', 'red', 'blue'])
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison of CNN vs. Traditional Methods')
plt.ylim(0, 100)
plt.show()
```

```python
lighting_conditions = ['Bright', 'Moderate', 'Dim', 'Low Light']
accuracy = [92, 90, 88, 85]

plt.figure(figsize=(8, 6))
plt.plot(lighting_conditions, accuracy, marker='o', color='purple')
plt.xlabel('Lighting Condition')
plt.ylabel('Accuracy (%)')
plt.title('Face Recognition Performance under Different Lighting Conditions')
plt.ylim(80, 100)
plt.grid(True)
plt.show()
```

```python
labels = ['Correctly Identified', 'Falsely Identified']
sizes = [92, 8]
colors = ['lightgreen', 'lightcoral']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Correctly vs. Falsely Identified Faces')
plt.show()
```
